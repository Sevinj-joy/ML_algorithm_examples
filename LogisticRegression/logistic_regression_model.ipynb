```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(data)

print("Original Data:\n", data)
print("Scaled Data:\n", scaled_data)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Visualization
plt.figure(figsize=(10, 5))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(data['Age'], data['EstimatedSalary'], color='blue')
plt.title("Original Data")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")

# Scaled data
plt.subplot(1, 2, 2)
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], color='orange')
plt.title("Scaled Data (MinMaxScaler)")
plt.xlabel("Scaled Age")
plt.ylabel("Scaled Estimated Salary")

plt.tight_layout()
plt.show()
```

```python
print(scaled_data.shape)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data['Age'], alpha=0.5, label='Age (Original)')
plt.hist(scaled_data[:, 0], alpha=0.5, label='Age (Scaled)')
plt.title("Age Distribution Before and After Scaling")
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 7))

ax1.plot(data['EstimatedSalary'], label='EstimatedSalary (Original)', color='black', linewidth=2)
ax1.set_ylabel('Original Salary')
ax1.set_title('Estimated Salary - Original vs Scaled')

ax2 = ax1.twinx()
ax2.plot(scaled_data[:, 1], label='EstimatedSalary (Scaled)', color='orange', linestyle='--', alpha=0.7)
ax2.set_ylabel('Scaled Salary (0 to 1)')

fig.legend(loc="upper left")
plt.show()

```

```python
normalized_salary = (data['EstimatedSalary'] - data['EstimatedSalary'].min()) / \
                    (data['EstimatedSalary'].max() - data['EstimatedSalary'].min())

plt.plot(normalized_salary, label='EstimatedSalary (Normalized)', color='blue')
plt.plot(scaled_data[:, 1], label='EstimatedSalary (Scaled)', color='red')
plt.legend()
plt.title('Estimated Salary (Normalized vs Scaled)')
plt.show()
```

```python
## Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression(random_state=0)
LR.fit(X_train,y_train)
```

```python
print(LR.predict(sc.transform([[30,87000]])))
```

```python
y_pred = LR.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

CM=confusion_matrix(y_test,y_pred)
print(CM)
accuracy_score(y_test,y_pred)
```

```python
## Visualising the Training set results

from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt

# Inverse transform the data
X_set, y_set = sc.inverse_transform(X_train), y_train

# Increase step size to avoid memory issues (can be further adjusted)
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 10))

# Create ListedColormap with correct usage
cmap = ListedColormap(['red', 'blue'])

# Generate contour plot
plt.contourf(X1, X2, LR.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = cmap)

# Set axis limits
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the training set results
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = cmap(i), label = j)

# Adding titles and labels
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()

# Show the plot
plt.show()


```

```python
## Visualising the Test set results

from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train

# Increase the step size to reduce the size of the grid
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 10))

# Use a smaller grid for predictions
plt.contourf(X1, X2, LR.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(['red', 'green']))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(['red', 'green'])(i), label = j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
